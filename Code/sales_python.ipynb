```python
car_sales_summary['brand']=car_sales_summary['name'].str.split().str[0]
car_sales_summary
```

```python
car_sales_summary['State'] = car_sales_summary['State'].str.upper()
car_sales_summary
```

```python
cars_sales_summary=car_sales_summary[['State','brand']]
cars_sales_summary
```

```python
cars_sales_summary_count=cars_sales_summary.groupby('State').agg(brands_sold=('brand','nunique')).reset_index()
cars_sales_summary_count
```

```python
final_sales= pd.merge(car_sales_region,cars_sales_summary_count, left_on='state',right_on='State')
final_sales
```

```python
final_sales=final_sales[['state','total_sales_value','total_cars_sold','brands_sold']]
final_sales
```

## Rank Calculation


From the metrics calculated above, it is important to rank each state based on its performance across the calculated metrics. The ranking criteria to be followed is:  

sales_rank : Higher the sales, better should be the rank. Ex. If state A sold $1000 worth of cars against state B which sold 800, then state A gets rank 1 and B gets rank 2.
cars_rank : Higher the number of cars sold, better should be the rank. Ex. If state A sold 100 cars against state B which sold 80, then state A gets rank 1 and B gets rank 2.
brands_rank : Higher the number of unique brands sold, better should be the rank.  Ex. If state A sold 10 unique brands against state B which sold 8, then state A gets rank 1 and B gets rank 2.

```python
final_sales['sales_rank']=final_sales['total_sales_value'].rank(method='dense' ,ascending=False)
final_sales['cars_rank']=final_sales['total_cars_sold'].rank(method='dense',ascending=False)
final_sales['brands_rank']=final_sales['brands_sold'].rank(method='dense',ascending=False)
final_sales
```

## State tier

It is now required to categorize states in different tiers. To achieve this purpose, following steps to be followed :



Calculate the average of all the rank columns and scale it finally between the range 0-10. (Use min-max scaling technique)

The criteria for categorizing states in different tiers is :
If scaled rank > 7, the state falls in Bottom Tier
If scaled rank is between 3 (exclusive) and 7, the state falls under Middle Tier
If the scaled rank is less than 3 (inclusive), the state is considered Top Tier


a1a442e6-95ff-4ac5-bd76-e70e247a872a_83d04ac6-cb74-4a96-a06a-e0d5442aa126_image.webp

```python
final_sales['avg_rank']=(final_sales['sales_rank']+final_sales['cars_rank']+final_sales['brands_rank'])/3
final_sales
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 10))
final_sales[['scaled_rank']] = scaler.fit_transform(final_sales[['avg_rank']])
final_sales
```

```python
def scaled(value):
    if value>7:
       return 'Bottom Tier'
    elif value>3 and value<7:
       return 'Mid Tier'
    else:
       return 'Top Tier'
```

```python
final_sales['state_tier']=final_sales['scaled_rank'].apply(scaled)
final_sales
```

```python
report=final_sales[['state','total_sales_value','total_cars_sold','brands_sold','scaled_rank','state_tier']]
report
```

```python
report.to_csv('report.csv',index=False)
```
